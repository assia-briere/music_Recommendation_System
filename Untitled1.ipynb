### 2. Using the Surprise Library (SVD)

We are using the SVD (Singular Value Decomposition) algorithm to predict user-item interactions.

Even though our dataset does not contain explicit ratings, we can treat the 'target' (user interaction)

as implicit feedback. The SVD model will predict how likely a user is to interact with a song,

and we can use these predictions to recommend songs based on the highest predicted scores.


```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add the steps of the recommendation flow as nodes
G.add_node("User-Item Matrix")
G.add_node("SVD")
G.add_node("Latent Factors")
G.add_node("Predicted Ratings")
G.add_node("Recommendation")

# Add edges to define the flow
G.add_edge("User-Item Matrix", "SVD")
G.add_edge("SVD", "Latent Factors")
G.add_edge("Latent Factors", "Predicted Ratings")
G.add_edge("Predicted Ratings", "Recommendation")

# Define position for each node for layout (more spaced out)
pos = {
    "User-Item Matrix": (0, 4),
    "SVD": (2, 2),
    "Latent Factors": (20, 2),
    "Predicted Ratings": (17, 3),
    "Recommendation": (26, 4)
}

# Draw the graph with node labels and arrows, adjust size and spacing
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=4000, 
        font_size=12, font_weight='bold', arrows=True, 
        arrowstyle='->', arrowsize=20)

# Adjust margins and display the plot
plt.margins(0.2)
plt.title('Recommendation Flow Diagram', fontsize=16)
plt.show()
```

#### a. Implementing SVD with Surprise 

```python
from surprise import Dataset, Reader
from surprise import SVD
from surprise.model_selection import cross_validate

# Create a dataframe with user_id, item_id, and target for Surprise
data_for_surprise = train_merged[['msno', 'song_id', 'target']]

# Define a reader with the rating scale (0 to 1 in your case)
reader = Reader(rating_scale=(0, 1))

# Load the data into Surprise's format
data = Dataset.load_from_df(data_for_surprise, reader)

# Use SVD for matrix factorization
algo = SVD()

# Evaluate the model using cross-validation (optional)
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)

# Train the model on the full dataset
trainset = data.build_full_trainset()
algo.fit(trainset)
```

```python

predicted_matrix = np.zeros(user_item_matrix.shape)

for i in range(user_item_matrix.shape[0]):
    for j in range(user_item_matrix.shape[1]):
        if np.isnan(user_item_matrix.values[i, j]):  # Predict only for missing values
            predicted_matrix[i, j] = algo.predict(i + 1, j + 1).est  # Predicting rating
```

```python
# Predict ratings for all items for the user
user_id = '++e+jsxuQ8UEnmW40od9Rq3rW7+A6996='  # Replace with the actual user ID
all_songs = train_merged['song_id'].unique()  # Get all unique song IDs
predictions = []

for song in all_songs:
    # Predict the rating for the user-song pair
    pred = algo.predict(user_id, song)
    predictions.append((song, pred.est))

# Convert to DataFrame for easier handling
predictions_df = pd.DataFrame(predictions, columns=['song_id', 'predicted_rating'])
```

```python
# Get songs the user has already interacted with
interacted_songs = train_merged[train_merged['msno'] == user_id]['song_id'].unique()

# Filter out already interacted songs
recommendations_df = predictions_df[~predictions_df['song_id'].isin(interacted_songs)]
```

```python
# Sort by predicted rating in descending order and select top N
top_n = recommendations_df.sort_values(by='predicted_rating', ascending=False).head(10)
```
